## FUNCTION TOOLS

```python
import os
os.environ["OPENAI_AGENTS_DISABLE_TRACING"] = "1"

import requests
from openai import AsyncOpenAI
from agents import Agent, Runner, OpenAIChatCompletionsModel, function_tool

client = AsyncOpenAI(
    api_key=OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY"),
    base_url="https://openrouter.ai/api/v1"
)

@function_tool
def get_weather(city: str) -> str:
    """Get current weather for a city."""
    api_key = "cb4714f596f46cad497226b3ef7c478c"
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    data = requests.get(url).json()
    return f"Temp: {data['main']['temp']}°C, Conditions: {data['weather'][0]['description']}, Humidity: {data['main']['humidity']}%"

weather_agent = Agent(
    name="Weather Agent",
    instructions="You are a weather assistant. Answer in a friendly paragraph.",
    model=OpenAIChatCompletionsModel(
        model="openrouter/free",
        openai_client=client
    ),
    tools=[get_weather]
)

weather_runner = await Runner.run(weather_agent, "Whats the weather in Lagos")
print(weather_runner.final_output)
```
